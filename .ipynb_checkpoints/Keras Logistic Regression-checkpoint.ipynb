```python
datasetPath = 'datasets/cancer/WPBC/WPBC.csv'
_CSV_COLUMNS = [
    'radius', 'perimeter', 'area', 'concave_points', 'col14', 
    'col16', 'col17', 'col24', 'col26', 'col27',
]

import tensorflow as tf
import tempfile
```

```python
def normalize(X):
    mean = np.mean(X)
    std = np.std(X)
    X = (X - mean)/std
    return X

```

```python
import pandas as pd
import numpy as np
```

```python
cancers = pd.read_csv(datasetPath)
cancers.head()
```

## labels
create two category column from outcome 0/1 column

```python
cancers['outcome'] = cancers['outcome'].astype('category')
y_data = {'outcome_R': np.where(cancers['outcome'] == 'R', 1, 0), 
          'outcome_N': np.where(cancers['outcome'] == 'N', 1, 0) }
y = pd.DataFrame(data=y_data)
y.head()
```

```python
X, Y = normalize(cancers[_CSV_COLUMNS]), y
```

```python
X.head()
```

```python
Y.head()
```

## Train-test division

```python
from sklearn.model_selection import train_test_split
seed = 7

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=seed)
```

```python
X_train.shape, X_test.shape
```

```python
from keras.models import Sequential
```

```python
from keras.layers import Dense, Activation
output_dim = nb_classes = 2
model = Sequential()
model.add(Dense(output_dim, input_dim=len(_CSV_COLUMNS), activation='softmax'))
```

```python
len(_CSV_COLUMNS)
```

## Compile the model

```python
model.compile(optimizer='sgd',
             loss='categorical_crossentropy',
              metrics=['accuracy'])
history = model.fit(X_train, y_train, 
                    batch_size=128,
                   nb_epoch=100,
                   verbose=1,
                   validation_data=(X_test, y_test))
score = model.evaluate(X_test, y_test, verbose=1)
print('Test score: ', score[0])
print('Test accuracy: ', score[1])
```
